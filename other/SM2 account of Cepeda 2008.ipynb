```python
import settings
difficulty = 0.95
settings.difficulty = 0.95
import forgetting_curve as fc
import matplotlib
import matplotlib.pyplot as plt
fc.get_current_pb_success(1, 2.5)
```

```python
def success_after_gap(gap, IRI):
    p1 = fc.get_current_pb_success(1, gap, difficulty=difficulty)
    p2 = fc.get_current_pb_success(gap, IRI, difficulty=difficulty)
    p3 = fc.get_current_pb_success(1, IRI, difficulty=difficulty)
    return p1*p2 + (1-p1)*p3 
```

```python
from operator import itemgetter
for IRI in [7, 35, 70, 350]:
    p = []
    gaps = range(1,IRI)
    couples = []
    for gap in gaps:
        s = success_after_gap(gap, IRI)
        p.append(s)
        couples.append((s,gap))
    print(str(IRI)+"-> best couple : "+str(max(couples, key=itemgetter(0))))
    plt.scatter(gaps,p,color='green')
    plt.show()
```
